```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
import cv2
```

```python
bl=cv2.imread('C:\\Users\\KIIT\\Desktop\\CB DS\\Collage Images\\bottom_left.jpg')
rgb_bl=cv2.cvtColor(bl,cv2.COLOR_BGR2RGB)
br=cv2.imread('C:\\Users\\KIIT\\Desktop\\CB DS\\Collage Images\\bottom_right.jpg')
rgb_br=cv2.cvtColor(br,cv2.COLOR_BGR2RGB)
tl=cv2.imread('C:\\Users\\KIIT\\Desktop\\CB DS\\Collage Images\\top_left.jpg')
rgb_tl=cv2.cvtColor(tl,cv2.COLOR_BGR2RGB)
tr=cv2.imread('C:\\Users\\KIIT\\Desktop\\CB DS\\Collage Images\\top_right.jpg')
rgb_tr=cv2.cvtColor(tr,cv2.COLOR_BGR2RGB)
ce=cv2.imread('C:\\Users\\KIIT\\Desktop\\CB DS\\Collage Images\\center.jpg')
rgb_ce=cv2.cvtColor(ce,cv2.COLOR_BGR2RGB)
```

```python
bl=cv2.resize(rgb_bl,(200,200))
br=cv2.resize(rgb_br,(200,200))
tl=cv2.resize(rgb_tl,(200,200))
tr=cv2.resize(rgb_tr,(200,200))
ce=cv2.resize(rgb_ce,(100,100))
```

```python
bce = cv2.copyMakeBorder(
    ce,
    top=10,
    bottom=10,
    left=10,
    right=10,
    borderType=cv2.BORDER_CONSTANT
)
btl = cv2.copyMakeBorder(
    tl,
    top=0,
    bottom=5,
    left=0,
    right=5,
    borderType=cv2.BORDER_CONSTANT
)
btr = cv2.copyMakeBorder(
    tr,
    top=0,
    bottom=5,
    left=5,
    right=0,
    borderType=cv2.BORDER_CONSTANT
)
bbl = cv2.copyMakeBorder(
    bl,
    top=5,
    bottom=0,
    left=0,
    right=5,
    borderType=cv2.BORDER_CONSTANT
)
bbr = cv2.copyMakeBorder(
    br,
    top=5,
    bottom=0,
    left=5,
    right=0,
    borderType=cv2.BORDER_CONSTANT
)
```

```python
col1=np.vstack([btl,bbl])
col2=np.vstack([btr,bbr])
```

```python
col=np.hstack([col1,col2])
bcol = cv2.copyMakeBorder(
    col,
    top=10,
    bottom=10,
    left=10,
    right=10,
    borderType=cv2.BORDER_CONSTANT
)
plt.imshow(bcol)
```

```python
l_img=bcol
s_img=bce
plt.imshow(s_img)
```

```python
x_offset=y_offset=155
l_img[y_offset:y_offset+s_img.shape[0], x_offset:x_offset+s_img.shape[1]] = s_img
```
